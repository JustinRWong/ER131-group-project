```python
def extract_flooding_measurements(_flood_meas, measurement='floods_historical'):
    ## create copy to not modify original
    flood_meas = _flood_meas.copy()
    
#     ## rename first colunmn to type
#     flood_meas = flood_meas.rename(mapper={'Unnamed: 0':'type'}, axis=1)
    
#     ## extract only the histrical measurements
#     flood_meas = flood_meas[flood_meas['type']=='floods_historical'].reset_index().drop('index', axis=1)
    
#     ## extract station names
#     station_names = flood_meas.columns[2:]
#     m = flood_meas.mean(axis=1)
    
#     for i, col in enumerate(flood_meas.drop(['type', 'Station Name'], axis=1)):
#         flood_meas.iloc[:,i] = flood_meas.iloc[:,i].fillna(m)
        
    
    ## filter by floods_historical, reserve year/Station Name, and drop the type, 'index', and Station Name columns
    flood_meas = flood_meas[flood_meas['type']==measurement].reset_index()
    reserved_year = flood_meas['Station Name']
    flood_meas = flood_meas.drop(['type', 'index', 'Station Name'], axis=1)
    
    
    ## fill Nans with mean
    flood_meas = flood_meas.apply(lambda row: row.fillna(row.mean()),axis=1)
    
    ## rename Station Name to year
    flood_meas.insert(0, 'year', reserved_year)
    return flood_meas
```

```python
def slice_and_dice_flood_stations(_flood_stations):
    ##  create a copy to not modify original
    flood_stations = _flood_stations.copy()
    
    ## melt to create a "cross product" across all columns, keeping the year and station info
    melted = pd.melt(flood_stations, id_vars=['Station Name'], value_vars=flood_stations.columns[2:])
    
    ## pivot the entire table  based on the Station Name, variable, and value
    melted = melted.pivot_table(index='variable', values='value', columns='Station Name')
    
    ## rename for neatness
#     melted.index = 'Station Name'
    return melted
```
