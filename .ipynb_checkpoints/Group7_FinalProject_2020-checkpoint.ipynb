This is the list of  libraries and cleaning/plotting code used later on.

````python
## Python libraries
import requests, os, time
from os import path
import math, random, csv

## Standard data science libraries
import numpy as np
import pandas as pd
import geopandas as gpd

## Plotting
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
import matplotlib.ticker as ticker
import matplotlib.lines as mlines

## sklearn and Cross Validation related
import sklearn
from sklearn.model_selection import train_test_split, KFold

## Models
import sklearn.linear_model as lm
from sklearn.linear_model import Ridge, Lasso
from sklearn.linear_model import RidgeCV, LassoCV
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor

## Model Measurements
from scipy import optimize
from scipy.stats import randint
from sklearn.metrics import mean_squared_error, r2_score

import warnings
warnings.filterwarnings('ignore')

'''
Wrapper for pd.read_csv to keep track of progress
'''
def extract_csv(directory, fn, log=False):
    fpath = directory + "/" + fn
    
    if log:
        print("\nOpening : " + fpath)
        
    df = pd.read_csv(fpath)
    
    if log:
        print("  Shape read: {s}".format(s=str(df.shape)))
    return df

'''
Given a directory, reads all the csv files and combines them into one dataframe.
'''
def extract_all_csv(csv_dir, log=False):
    files = os.listdir(csv_dir)
    
    combined = pd.DataFrame()
    for fn in files:
        if fn[-4:] == '.csv':
            extracted = extract_csv(csv_dir, fn, log=log)

            combined = pd.concat([combined, extracted])
    return combined

'''
Wrapper for extracting geopandas dataframe. Must open a shp file
'''
def extract_gpd(directory, fn, log=False):
    fpath = directory + "/" + fn + "/" + fn + ".shp"
    if log:
        print("\nOpening : " + fpath)
    gpd_df = gpd.read_file(fpath)
    
    if log:
        ## Print shape and each unique geometry type
        print("  Shape: {s}".format(s=gpd_df.shape))
        print("  Geomtry types")
        [print("    -" + t) for t in gpd_df.geometry.type.unique()]
    return gpd.read_file(fpath)

'''
_data: a datframe to create scatter plots
y: the target variable
cols: the columns that you want to create a scatterplot with y

Plots all the scatter plots between y and each indivudal column specified in `cols`
'''
def standard_bulk_scatter_plot(_data, cols, y, n_cols=2, square_size=15, fsize=13, ratio=.8):
    n = len(cols) ## this determines how many plots we need to make
    
    rows = math.ceil(n / n_cols) ## this determines how many rows of n_cols we will plot
    fig = plt.figure(figsize=(square_size, square_size*ratio))
    gs = fig.add_gridspec(square_size, n_cols)
    
    switch = 0 ## flips betwen left and right plot on each row
    for i in range(n):
        ## move to other plot
        switch = i % n_cols
        
        c = cols[i]
        r = i // n_cols ## the current row of this iteration
        ax1 = fig.add_subplot(gs[r, switch])
        
        ## scatterplot
        sns.scatterplot(data=_data, x=c, y=y)
        
        ## chart design
        plt.xlabel(c + " Count")
        plt.ylabel(y + " in Tons Per Day")
        plt.title('{y} vs {c}'.format(y=y, c=c))
        
        plt.tight_layout()
        

'''
Designed to be written such that `x` is a time variable
and plots multiple columns specified by `y` in additional subplots
'''
def standard_bulk_historical_line_plot(_data, x, y, xlabel, hue, n_cols=2, square_size=15, fsize=13, loc=0, ratio=.8, bbox_to_anchor=(1.05, 1)):
    assert type(x) == str, "x is type {t}".format(t=type(x))
    assert type(y) == list, "y is type {t}".format(t=type(y))
    
    n = len(y) ## this determines how many plots we need to make
    
    rows = math.ceil(n / n_cols) ## this determines how many rows of n_cols we will plot
    fig = plt.figure(figsize=(square_size, square_size*ratio))
    
    gs = fig.add_gridspec(rows, n_cols)
    
    switch = 0 ## flips betwen left and right plot on each row
    for i in range(n):
        ## move to other plot
        switch = i % n_cols
        
        c = y[i] ## the ith column to plot
        r = i // n_cols ## the current row of this iteration
        ax1 = fig.add_subplot(gs[r, switch])
        
        ## lineplot
#         print(c, y)
        sns.lineplot(data=_data, 
                     x=x, 
                     y=c, 
                     hue=hue)
        
        ## chart design
        plt.title('{x} over {y}'.format(x=c, y=xlabel), size=fsize+2)
        ax1.legend(loc=loc, bbox_to_anchor=bbox_to_anchor)
        plt.xlabel(xlabel,size=fsize)
        plt.ylabel(c,size=fsize)
        
        
        plt.tight_layout()


'''
`_data`: a datframe to create scatter plots
`y`: the target variable
`cols`: the columns that you want to create a scatterplot with y
`n_cols`: the number of vertical columns to bulk plot
`square_size`: the figure size
`fsize`: font size of the axes. Title font will be fsize+2

Plots all the heatmaps plots between y and each indivdual column specified in `cols`
'''
def standard_bulk_geoplot(_data, cols, titles, n_cols=2, square_size=15, fsize=13, ratio=.8):
    assert len(cols) == len(titles)
    n = len(cols) ## this determines how many plots we need to make
    
    fig = plt.figure(figsize=(square_size, square_size*ratio))
    
    rows = math.ceil(n / n_cols) ## this determines how many rows of 2 we will plot
    gs = fig.add_gridspec(rows, n_cols)
    
    switch = 0 ## flips betwen left and right plot on each row
    for i in range(n):
        ## move to other plot
        switch = i % n_cols
        
        c = cols[i]
        r = i // n_cols ## the current row of this iteration
#         print(r,switch)
        ax1 = fig.add_subplot(gs[r, switch])
        
        ## geoplot
#         print(c, y)
        _data.plot(ax=ax1, column=c, legend=True)

    
#         ax1.axis('on') # You can optionally omit the axes

        # Show a title
        
        plt.title(titles[i], size=fsize+2)
        plt.xlabel('Longitude',size=fsize)
        plt.ylabel('Latitude',size=fsize)
        
        plt.tight_layout()
        

'''
Displays a corrrelation chart for the  specified features in the edf
'''
def corr_vis(df, features, meas='TOG'):
    ## Create a blank dataframe that calculates all the correlations to TOG
    vis = pd.DataFrame()
    vis['feature'] = features
    vis['spearman'] = [df[f].corr(df[meas], 'spearman') for f in features]
    #[print(str(features[f]) + ": " + str(spr['spearman'][f])) for f in range(len(features))]
    
    vis = vis.sort_values('spearman')
    plt.figure(figsize=(6, 0.25*len(features)))
    ax = sns.barplot(data=vis, y='feature', x='spearman', orient='h')
    
    ## Chart design
    ax.set(xlabel='correlation')
    ax.set(title='Correlation of all features to {meas}'.format(meas=meas)) 
    return vis
````

The data comes from US Census Bureau, Department of Commerce. This dataset provides spatial data products with identified and organized locations of potential methane (CH4) emitting facilities and infrastructure spanning the State of California. These data products form a GIS-based mapping database designed to address shortcomings in current CH4 source inventories and is known as Vista California (Vista-CA). Vista-CA consists of detailed spatial maps for facilities and infrastructure in California that are known or expected sources of CH4 emissions and illustrates the spatial distribution of potential CH4 sources. 

We see that Vista CA provides spatial data products with identified and organized locations of potential methane (CH4) emitting facilities and infrastructure spanning the State of California. However, it is missing the actual measurement.

Vista-CA spatial data sets were created utilizing an assortment of publicly available data sources from local, state, and federal agencies for the years 2005 to 2019. The final Vista-CA database contains over 230,000 entries, which are presented as fifteen CH4 emitting infrastructure maps. The database was used to support flight planning and source attribution for the California Methane Survey project.

The Vista-CA data product is a key tool for CH4 emissions research and mitigation efforts; by (1) mapping areas of CH4 emitting infrastructure, (2) identifying targets for CH4 surveys, and (3) enabling interpretation of atmospheric observations, including source attribution, and comparison of measured emissions to permitted or reported emissions. Combined with atmospheric observations, Vista enables systematic study of CH4 emission sources.

### Data Cleaning Specifics
The following section will combine our data sources together to create a unified dataframe so that the prediction problem can be tackled. This giant unified dataframe `aggregated_datasets_gdf` is a geopandas dataframe that aggregates all our datasets.

Because we used many diffeerent datasets, two common task we had to do across all our datasets was:
- Use a consistent `County` name o use a consistent county naming convention. 
    - Our `counties` data included the centroids of each county from the coordinates defined by (INTPTLAT, INTPTLON). However, the counties across different datasets were spelled differently! For example, "Alameda",  "Alameda County", and "ALAMEDA" all refer to the same county but would not be able to joined properly. Nearly every dataset had to be cleaned to account for this.
- Drop irrelevant columns
    - To merge the datasets together, we had to get rid of many single-dataset specific columns, many of which were String data. 

To convert county names easily, we decided to call counties by all upper case and ignore " county" in the name. For example, our county name would have "ALAMEDA"


#### 1. California Counties
No significant modifications was done to this dataset, except dropping certain rows and renaming the `Longitude` and `Latitude` columns into `Centroid_Longitude` and `Centroid_Latitude` to prevent future duplicate column names.


#### 2. Cleaning Vista CA
Since there are many shape files for the different Vista CA emissions sources, we want to extract all these individual files and put them into one dataframe all togeher. This will become the `combined_vista_ca` dataframe, which isn't shown.

The Vista CA indicates the locations of different emissions source stations in California. Since the some columns across the multiple CA Vista files represent the same thing, we need to combine and clean the columns. Below are the columns that represent the same thing:


`Latitude`<-- ['Latitude', 'LATITUDE', 'latitude']

`Longitude`<-- ['Longitude', 'LONGITUDE', 'longitude']

`Vista_Date`<--  ['VistaDate', 'Vista_Date']

`State`<-- ['State', 'STATE']

`City`<-- ['City', "CITY"]

`Source`<-- ['Source', 'SOURCE', 'source']

We started with a blank DataFrame. Then each Vista CA dataset was extracted from geoepandas, converted to the correct coordinate refernece system, and concatenated to the initial DataFrame. Lastly, we renamed the `Longitude` and `Latitude` columns into `Station_Longitude` and `Station_Latitude` to prevent duplicate column names.

Because we seek to predict atmospheric carbon emissions, county borders may not necessarily reflect the true air quality of a station. Thus, we want to minimize the distance from a station to a county centroid. To do so, we wrote two functions `find_distance` and `find_closest_county`, inspired from a previous assignment, which would determine the closest county centroid to a Vista CA facility.

While county borders are clearly defined, there are some cases where the distance from a station to any county centroid is closer to a neighboring county's centroid than the station's true county's centroid. We called these `Displaced Stations` because it is "displaced" from its nearest county centroid, but defined as a different county. There were 261 displaced stations. 

We also grouped the numbere of each facility in a particular county as features in an aggregated DataFrame.


#### 3. Cleaning EMSEIC: Emissions by Emission Inventory County
We also ensure we are using counties_wgs84 so that we are using the correct coordinatee reference system. Since each row was a measurement within a county for different years, we grouped by year and county so that the resulting dataframe encompass the estimated emissions per county per year.

First thing that we neeed to do is standarize our data.

````python
"""
This function returns four dataframes containing the testing and training X and y values used in land-use regression.
It standardizes the X variables prior to splitting the data.
Input: df, a Pandas dataframe with all of the fields in the land-use regression dataset; 
    cols_to_drop, a list of the names (strings) of the columns to drop from df in order to obtain the feature variables.
    y_col, a column name (as a string) of df that represents the response variable
    test_prop, a float between 0 and 1 indicating the fraction of the data to include in the test split
    rand_seed, an integer, used to define the random state
Returns: X_train, X_test, y_train, y_test, four dataframes containing the training and testing subsets of the 
feature matrix X and response matrix y
"""
def get_X_y_stnd(df, y_col, test_prop, rand_seed, drop_cols = []):
    if y_col not in drop_cols:
        drop_cols.append(y_col)
    
    assert y_col in df.columns, "{y} column is not in the dataframe".format(y=y_col)
    X = df.drop(drop_cols, axis=1)
    y = df[[y_col]]
    
    scaler = StandardScaler()
    scaler.fit(X)
    x_trans = scaler.transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(x_trans, y ,test_size = test_prop, random_state = rand_seed)
    
    return X_train, X_test, y_train, y_test
````

````python
def calculate_mses(alphas, Model):
    """
    Input:
        alphas (array): contains floats of various alpha values
        Model (sklearn model): the type of sklearn model on which to fit the data
    Output:
        an array of floats containing the mean-squared-errors from the predictions
    """
    mses = []

    for a in alphas:
        model = Model(alpha=a)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mses.append(mean_squared_error(y_pred, y_test))
        
    return mses
````

#### PQ2 - Model #4 and #5: LassoCV and RidgeCV Regression


Comparing Ridge and Lasso, Lasso performs more consistently and has lower MSEs than ridge regression. While our funciton `model_cv_mse` does cross validation with k=3, we want to compare with sklearn's RidgeCV and LassoCV. 

````python
"""
This function calculates the MSE resulting from k-fold CV using Lasso or Ridge regression performed on a training subset of 
X and y for different values of alpha.

Also times of how long it took for the model to train and predict for easy comparison.

Inputs: 
    Model (sklearn model): the type of sklearn model with which to fit the data - LinearRegression, Ridge, or Lasso
    X: the set of features used to fit the model
    y: the set of response variable observations
    alphas: a list of penalty parameters
    k: number of folds in k-fold cross-validation
Returns:
    mses: a list containing the mean squared cross-validation error corresponding to each value of alpha
"""
def model_cv_mse(Model, X, y, alphas, k = 3):
    start_time = time.time()
    mses = np.full((k, len(alphas)), np.nan) # initialize array of MSEs to contain MSE for each fold and each value of alpha
        
    kf = KFold(k, random_state=9,shuffle = True) # get kfold split
    
    #makes y a series, so can index

    
    fold = 0
    for train_i, val_i in kf.split(X):
        # get training and validation values
        X_f_train = X[train_i]
        X_f_val = X[val_i]
        y_f_train = y.iloc[train_i,0]
        y_f_val = y.iloc[val_i,0]
    
        
        for i in range(len(alphas)): # loop through alpha values
            model = Model(alpha = alphas[i]) # initialize model

            model.fit(X_f_train,y_f_train) # fit model
            
            
            y_pred = model.predict(X_f_val) # get predictions
            
            # save MSE for this fold and alpha value
            mses[fold, i] = mean_squared_error(y_f_val, y_pred)
        
        fold += 1
    
    average_mses = np.full((len(alphas), 1), np.nan) # get average MSE for each alpha value across folds
    for i in range(len(alphas)):
        average_mses[i,:] = np.mean(mses[:, i])
    
    print("Finished Training model_cv_mse in - %s seconds -. Returning..." % (time.time() - start_time))
    return average_mses, model.coef_

def plot_mse(_range, model_CV, model_name):
    plt.plot(_range, model_CV, label=model_name)
    plt.title('Try to find the Alpha value for '+ model_name +' min MSE')
    plt.xlabel('alpha')
    plt.ylabel('MSE')
````

LassoCV and RidgeCV

We decided implement CV in two ways, one using sklearn's `LassoCV` and `RidgeCV` and our own written `model_cv_mse` function. We used thee sklearn models to find the optimal learning rate `model_alpha_opt`. We then use the values that are returned as benchmarks to compare our models.

````python
'''
Generalized sklearn modelCV initialization, fit, and predicting
and returns the optimal alpha from sklean module. 

Also times of how long it took for the model to train and predict for easy comparison.
'''
def sklearn_modelCV_process(Model, X_train, X_test, y_train, y_test, _alphas, kf_random_state=2020, kf_shuffle=True):
    start_time = time.time()
    ## Set up kfold
    kf = KFold(len(X_train), random_state= kf_random_state, shuffle= kf_shuffle) # get KFold cross-validation selector object
    
    ## Initalize and train model
    modelcv = Model(cv=kf, alphas=_alphas) # pass CV object and list of alphas to initialize the Model
    modelcv.fit(X_train,y_train) # fit modelcv on training data
    
    print("Finished Training sklearn Model in - %s seconds -" % (time.time() - start_time))
    
    ## obtain the optimal alpha
    model_alpha_opt = modelcv.alpha_ # get optimal alpha value
    
    ## use the model to predict X_test
    y_pred_modelcv = modelcv.predict(X_test) # get test predictions using RidgeCV model
    
    ## Calculate MSE of tthe model
    lassocv_mse_sklearn = mean_squared_error(y_test, y_pred_modelcv) # get MSE of modelcv
    
    speed = time.time() - start_time
    print("Finished ModelCV sklearn Proceess in - %s seconds -" % (speed))
    return modelcv, model_alpha_opt, y_pred_modelcv, lassocv_mse_sklearn, speed

````

````python
# Our learning rate range for lasso
lasso_alphas = np.arange(0.0001,0.06, 0.001)

## Initialize, train, and test out LassoCV.
# We use LassoCV from sklearn to get optimal alpha.
lassocv, lasso_alpha_opt, y_pred_lassocv, lassocv_mse_sklearn, lassocv_speed = sklearn_modelCV_process(LassoCV, 
                                                                                        X_train, 
                                                                                        X_test, 
                                                                                        y_train, 
                                                                                        y_test, 
                                                                                        lasso_alphas, 
                                                                                        kf_random_state=2020, 
                                                                                        kf_shuffle=True)

# LASSO CV using our own cross validation function
lassocv_mse_coef = model_cv_mse(Lasso, X_train, y_train, alphas = lasso_alphas)
lassocv_mse = lassocv_mse_coef[0]
lassocv_coef = lassocv_mse_coef[1]
````

````python
# # Our learning rate range for Ridge
ridge_alphas = np.arange(0.0001,0.1,.001)

## Initialize, train, and test out RidgeCV.
# We use RidgeCV from sklearn to get optimal alpha.
ridgecv, ridge_alpha_opt, y_pred_ridgecv, ridgecv_mse_sklearn, ridgecv_speed = sklearn_modelCV_process(RidgeCV, 
                                                                                        X_train, 
                                                                                        X_test, 
                                                                                        y_train, 
                                                                                        y_test, 
                                                                                        lasso_alphas, 
                                                                                        kf_random_state=2020, 
                                                                                        kf_shuffle=True)

# LASSO CV using our own cross validation function
ridgecv_mse_coef = model_cv_mse(Ridge, X_train, y_train, alphas = ridge_alphas)
ridgecv_mse = ridgecv_mse_coef[0]
ridgecv_coef = ridgecv_mse_coef[1]
````

````python
'''
Wrapper function to compare Lasso and Ridge CV
Takes in lists for various model compairson measurements and plots 3 plots
Topleft: the first model MSE vs alphas and includes optimal alpha, test MSE, and time taken for the training
Topright: the second model MSE vs alphas and includes optimal alpha, test MSE, and time taken for the training
Bottom: plots both models MSE vs alphas on the same plot to compare with the same scale
'''
def visualize_lassocv_vs_ridgecv(optimal_alphas, test_cv_mses, model_coefs, r2_scores, speeds, ranging, model_title, alphas, sklearn_mses):
    ## Plot lassoCV vs ridgeCV mse and alphas to compare which does better 
    fig = plt.figure(figsize = (15,10), constrained_layout=True)
    gs = fig.add_gridspec(2, 2, figure=fig)

    for alpha_opt, cv_mse, m_coef, r_square, speed, i, title, alphs, sk_mses in zip(optimal_alphas, test_cv_mses, model_coefs, r2_scores, speeds, ranging, model_title, alphas, sklearn_mses):
        ## Plot only single model individually
        ax = fig.add_subplot(gs[0, i])
        plot_mse(alphs, cv_mse, title)
        
        print("{model} optimal alpha: {v}".format(v=alpha_opt, model=title))
        print("{model} Test MSE: {v}".format(v=sk_mses, model=title))
        print('{model} R^2: {v}'.format(v=r_square, model=title))
        print('{model} Speed: {v} seconds\n'.format(v=speed, model=title))
    
        cv_text = '{m}\n'.format(m=title) + \
                'Optimal alpha       : {v}\n'.format(v=alpha_opt) + \
                "Test {m} MSE : {v}\n".format(v=sk_mses, m=title) + \
                'Speed                    : {v} seconds'.format(v=speed)
        
        plt.text(.25, 0.5, cv_text, horizontalalignment='left',
             verticalalignment='center', transform=ax.transAxes)
        plt.legend()
#         plt.tight_layout()
        
    ## Plot both LassoCV and RidgeCV to compare on same plot
    ax_bot = fig.add_subplot(gs[1, :])
    
    plot_mse(alphas[0], test_cv_mses[0], model_title[0])
    plot_mse(alphas[1], test_cv_mses[1], model_title[1])
    plt.title('Comparing Ridge and Lasso MSE vs alpha ')
    plt.legend()
    plt.tight_layout()
    
````

#### PQ 3 - Model #3: Gradient Boosting Regressor and Randomized Search CV

Gradient Boosting Regressor (GBR) is an ensemble method where a new tree is chosen to minimize the residual of an existing aggregation of trees. The number of boosts or the parameter `n_estimators` can be tuned using cross validation. Gradient Boosting Regressor builds an additive model in a forward stage wise-fashion. Each iteration the regression tree is fit on the negative gradient of the given loss function.

````python
'''
Downloads a csv file from `source` and 
stores it in the specified `fdirectory` with the name `fn`.
'''
def download_to_directory_as_fn(fdirectory, fn, source):
    ## write to this specific path
    full_fn = "{fdir}/{csv}".format(fdir=fdirectory, csv=fn)
    
    if fn not in fdirectory:
        r = requests.get(source)
        open(full_fn, 'wb').write(r.content)
        print("Processed {fn} to {dest} <-- {url}".format(fn=fn, url=source, dest=fdirectory))
    else:
        print("Already downloaded {fn}!".format(fn=fn))
        
'''
Downloads California Air Resources Board EMSEIC data
filtered by year and county. Default  to ALL counties.
`year`: the year for the desired EMSEIC data
`county`: a number from 1-58, which specifies which county to download EMSEIC data from. 
            `county` defaults to zero, which will download all 58 counties.
'''
def download_all_arb_CA_counties_emseic_by_year_and_county(year, county=0, fdir=''):
    ## replacing year for parameterized php request
    base = "https://www.arb.ca.gov/app/emsinv/2017/emsbyeic.csv?F_YR={year}&F_DIV=-4&F_SEASON=A&SP=SIP105ADJ&SPN=SIP105ADJ&F_AREA=CO&F_COAB=&F_CO=".format(year=year)
    ## Check if emseic/{year} currently exists as a folder and if not, create it
    if fdir == '':
        fdir = "data/emseic/{year}".format(year=year)
    else:
        fdir = "data/emseic/{custom}".format(custom=fdir)
    files = []
    if path.isdir(fdir):
        files = os.listdir(fdir)
    else:
        os.mkdir(fdir)
        print("Directory '%s' created" %fdir) 
    
    ## if county is 0 (default), and 
    if county == 0:
        ## iterate through all 58 counties 
        for i in range(1, 59):
            ## replacing county number for parameterized php request
            county_download_url = base + str(i) 
            
            ## specify the location to download the file
            csv_file = "emseic_{year}_{i}.csv".format(i=i, year=year)
            fn = "{fdir}/{csv}".format(fdir=fdir, csv=csv_file)
            
            ## download the data for that year
            download_to_directory_as_fn(fdir, csv_file, county_download_url)
    else:
        ## just download a single county
        county_download_url = base + str(county)
        csv_file = "emseic_{year}_{c}.csv".format(c=county, year=year)
        fn = "{fdir}/{csv}".format(fdir=fdir, csv=csv_file)
        download_to_directory_as_fn(fdir, csv_file, county_download_url)
            
'''
Wrapper to download CARB EMSEIC data in a year range and specific county.
For example, 
```
### Run this to download all ARB emissions for all CA counties between 2000 to 2012
download_all_arb_CA_counties_emseic_by_yearrange_and_county(2000, 2012)
```
'''
def download_all_arb_CA_counties_emseic_by_yearrange_and_county(begin, end, county=0, fdir=''):
    for y in range(begin, end+1):
        ## if downloading all counties, download all counties for a single year to take advantage of some caching
        if county == 0:
            for c in range(1, 59):
                download_all_arb_CA_counties_emseic_by_year_and_county(y, county=c)
        else:
            download_all_arb_CA_counties_emseic_by_year_and_county(y, county=county, fdir=fdir)

'''
Combines  and cleans a dataframe by clearing Nans on the specified `columns_to_combine`
and concatenates the results ina new column specified by `combined_col_name`.
This then drops the `columns_to_combine` and returns the dataframe.

This is can optionally mutate the dataframe inplace.
'''
def combine_and_clean(_df, combined_col_name, columns_to_combine, inplace=False):
    assert type(columns_to_combine) == list, "`columns_to_combine` must be a list"
    assert type(combined_col_name) == str, "`combined_col_name` must be a list"
    ## Make  sure each columns_to_combine is in actually a column in our dataframe
    df_columns = _df.columns
    for input_col in columns_to_combine:
        assert input_col in df_columns, "Make sure you include a column that exists in the dataframe"
    
    df=_df
    if not inplace:
        df = _df.copy()
        
    first = columns_to_combine[0]
    firstIndexOccurence= df.columns.get_loc(first)
    
    clean_replacement = df[columns_to_combine].fillna('').astype(str).agg(','.join, axis=1)
    clean_replacement = [val.strip(',') for val in clean_replacement]
    df = df.drop(columns_to_combine, axis=1)
    df.insert(firstIndexOccurence, combined_col_name, clean_replacement)
    return df

'''
Wrapper for `combine_and_clean` that combines a series of multiple columns together.
'''
def bulk_clean_and_combine(mapper, _data, log=False):
    data = _data.copy()
    
    for k, v in mapper.items():
        if log:
            print(k,v)
            
        data = combine_and_clean(data, k, v)
        
        if log:
            print("  Finished processing {k}".format(k=k))
            
    return data
````
